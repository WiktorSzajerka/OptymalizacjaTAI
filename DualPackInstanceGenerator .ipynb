```python
import math

class RandomNumberGenerator:
    def __init__(self, seedVaule=12345):
        self.__seed=seedVaule
    def nextInt(self, low, high):
        m = 2147483647
        a = 16807
        b = 127773
        c = 2836
        k = int(self.__seed / b)
        self.__seed = a * (self.__seed % b) - k * c;
        if self.__seed < 0:
            self.__seed = self.__seed + m
        value_0_1 = self.__seed
        value_0_1 =  value_0_1/m
        return low + int(math.floor(value_0_1 * (high - low + 1)))
    def nextFloat(self, low, high):
        low*=100000
        high*=100000
        val = self.nextInt(low,high)/100000.0
        return val


randomGenerator = RandomNumberGenerator()
```

```python
def write_random_table_to_dat(file_name, objectsNumber, seed=12345):
    rng = RandomNumberGenerator(seed)

    packSize = 0;

    with open(file_name, "w", encoding="utf-8") as f:
        f.write(f"objectsNumber = {objectsNumber}; \n")
        f.write("values = [\n")
        totalSize = 0;
        for r in range(objectsNumber):
            row_values = []
            combinedSize = 0;
            for c in range(3):
                size = rng.nextFloat(1, 10);
                if c != 0 :
                    combinedSize += size;
                row_values.append(str(size))
            totalSize += combinedSize/2;
            f.write("[" + ",".join(row_values) + "]")
            if r < objectsNumber - 1:
                f.write(",\n")
            else:
                f.write("\n")
        f.write("];\n")
        randomSize = rng.nextInt(math.floor(totalSize/4), math.floor(totalSize/2));
        f.write(f"size = {randomSize}; \n")
```

```python
for objectsNumber in range (10, 40, 5):
    write_random_table_to_dat(f"values_{objectsNumber}.dat", objectsNumber, seed=12345)
```
