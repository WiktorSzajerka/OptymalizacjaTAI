```python
import math

class RandomNumberGenerator:
    def __init__(self, seedVaule=12345):
        self.__seed=seedVaule
    def nextInt(self, low, high):
        m = 2147483647
        a = 16807
        b = 127773
        c = 2836
        k = int(self.__seed / b)
        self.__seed = a * (self.__seed % b) - k * c;
        if self.__seed < 0:
            self.__seed = self.__seed + m
        value_0_1 = self.__seed
        value_0_1 =  value_0_1/m
        return low + int(math.floor(value_0_1 * (high - low + 1)))
    def nextFloat(self, low, high):
        low*=100000
        high*=100000
        val = self.nextInt(low,high)/100000.0
        return val


randomGenerator = RandomNumberGenerator()
```

```python
def write_random_table_to_dat(file_name, dimension, low, high, seed=12345):
    rng = RandomNumberGenerator(seed)

    with open(file_name, "w", encoding="utf-8") as f:
        f.write(f"dimension = {dimension}; \n")
        f.write("values = [\n")
        for r in range(dimension):
            row_values = []
            for c in range(dimension):
                row_values.append(str(rng.nextInt(low, high)))
            f.write("[" + ",".join(row_values) + "]")
            if r < dimension - 1:
                f.write(",\n")
            else:
                f.write("\n")
        f.write("];\n")
```

```python
for dimension in range (10, 100, 10):
    write_random_table_to_dat(f"values_{dimension}.dat", dimension, low=1, high=50, seed=12345)

```
